```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('/home/adrianwerle/Descargas/perro.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen)
```

```python
imagen.shape
```

```python
imagen2 = cv2.resize(imagen, (600,500))
```

```python
plt.imshow(imagen2)
```

```python
ratio_ancho = 0.5
ratio_alto = 0.5
```

```python
imagen2 = cv2.resize(imagen2, (0,0), imagen2, ratio_ancho, ratio_alto)
```

```python
plt.imshow(imagen2)
```

```python
imagen2.shape
```

```python
imagen2 = cv2.flip(imagen2, 1)
```

```python
plt.imshow(imagen2)
```

```python
imagen2 = cv2.flip(imagen2, 0)
plt.imshow(imagen2)
```

```python
imagen2 = cv2.cvtColor(imagen2, cv2.COLOR_RGB2BGR)
```

```python
cv2.imwrite('/home/adrianwerle/Descargas/perro_al_reves.jpg', imagen2)
```
